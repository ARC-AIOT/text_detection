```python
import numpy as np
import matplotlib.pyplot as plt

import cv2
img = cv2.imread("5.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.resize(img,(640, 480),interpolation=cv2.INTER_CUBIC)
print(img.shape)
plt.imshow(img)
plt.show()
```

```python
img_arr = list(np.ndarray.flatten(img))
samples = ""
wrapped_arr = [img_arr[i:i + 20] for i in range(0, len(img_arr), 20)]
for sub_arr in wrapped_arr:
    samples += "\t\t" + str(sub_arr)[1:-1] + ",\n"

samples_file = open("test_samples.cpp", "w")
samples_file.write(samples);
samples_file.close()
```
