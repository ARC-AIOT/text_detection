```python
import numpy as np
# 算row的標準差, 可決定幾個pixel算一次, 最後加總
def abs_std(arr):
  avg = np.mean(arr)
  sum = 0
  for i in range(0, arr.shape[0]):
    sum += abs(arr[i]-avg)
  return sum/arr.shape[0]
  
def local_SD(row, l):
  sum = 0
  for i in range(0, int(row.shape[0]/l)):
    end = i+l if i+l <= row.shape[0] else row.shape[0]
    sum += abs_std(row[i : end])
  return sum/int(row.shape[0]/l)

# 算好標準差後, 找出這些標準差的local minimum, 且標準差低於threshold的row, 此row即為空白row
# 設定stride的原因是, 若運算量太大可以把stride調2試試看 (實測後發現=1即可)
def find_local_min(std, stride):
  avg = np.mean(std)
  out = []
  for i in range(1, len(std)-1):
    if std[i] < avg and std[i-1] > std[i] and std[i] < std[i+1]:
      out.append(i*stride)
  return out
```

```python
import os
import cv2

data_path = '/content/img/'
for file in os.listdir(data_path):
  image = cv2.imread(data_path+file)
  image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  arr_std = []
  for i in range(0, image.shape[0], 1):
    # 經過實驗發現每3個做一次標準差效果好
    arr_std.append(local_SD(image[i], 3))
  line = find_local_min(arr_std, 1)
  #找出前10個間隔大的line, 這些line中有文字的可能很大
  diff = np.array(line[1:len(line)])-np.array(line[0:len(line)-1])
  diff = list(enumerate(diff))
  diff = sorted(diff, key = lambda s: s[1], reverse = True)
  for i in range(0, 10):
    img = image[line[diff[i][0]]:line[diff[i][0]+1]][:]
    cv2.imwrite('/content/drive/MyDrive/MyDataset/text_recognition/img/'+file[:-4]+'_'+str(i+1)+'.jpg', img)
  print(file+' OK')
```
